```python
# Ques 4: dataset-- titanic
#     Set up a grid to plot survival probability against several variables like class, sex, who and alone.
#     Then Draw a seaborn pointplot onto each Axes
#         ## output should be as shown below-
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.pointplot(x='class', y='survived', data=titanic, ax=axes[0, 0], ci=None)
axes[0, 0].set_title('Survival Probability vs Class')
sns.pointplot(x='sex', y='survived', data=titanic, ax=axes[0, 1], ci=None)
axes[0, 1].set_title('Survival Probability vs Sex')
sns.pointplot(x='who', y='survived', data=titanic, ax=axes[1, 0], ci=None)
axes[1, 0].set_title('Survival Probability vs Who')
sns.pointplot(x='alone', y='survived', data=titanic, ax=axes[1, 1], ci=None)
axes[1, 1].set_title('Survival Probability vs Alone')

plt.tight_layout()
plt.show()
```

```python
# Ques 5: dataset-- diamonds
#     Generate the distribution of prices with respect to color for each cut quality.
#     ## output should be as shown below
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("C:\\Users\\erpra\\OneDrive\\Desktop\\SHIV SEABORN TASK\\diamonds.csv")
g = sns.FacetGrid(df, col='cut', col_wrap=3, height=4, aspect=1.5)
g.map_dataframe(sns.histplot, x='price', hue='color', multiple='stack', palette='Set1', kde=True)
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Distribution of Diamond Prices with Respect to Color for Each Cut Quality')
plt.show()
```

```python
# Ques 9: dataset-- titanic
#     Draw a nested barplot to show survival for class and sex
#     ## output should be as shown below
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

survival_rates = titanic.groupby(['pclass', 'sex'])['survived'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=survival_rates, x='pclass', y='survived', hue='sex', palette='pastel')
plt.title('Survival Rates by Class and Sex')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)  # Set y-axis limit from 0 to 1 for survival rate
plt.legend(title='Sex')
plt.show()
```

```python
# Ques 10: dataset-- fmri()
#     Create line plot to show the responses for different events and regions
#     ## output should be as shown below
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset('fmri')

plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region', markers=True, dashes=False)
plt.title('FMRI Signal Response by Event and Region')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Ques 11: dataset-- iris
#     Create a scatter plot between sepal length and sepal width categorised by species.
#             ## output should be as shown below
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1', s=100)
plt.title('Sepal Length vs Sepal Width categorized by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

```python
# Ques 12: dataset-- iris
#     Visualize the kernel density estimate of petal length for each species.
#     ## output should be as shown below
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris, x='petal_length', hue='species', common_norm=False, palette='Set1', alpha=0.7)
plt.title('Kernel Density Estimate of Petal Length by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.legend(title='Species')
plt.show()
```

```python
# Ques 13: dataset-- iris
#     Study the relationship between sepal length and sepal width for the iris dataset by creating a joint plot.
#     ## output should be as shown below
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
sns.set(style="white", color_codes=True)
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='scatter', height=7, marginal_kws=dict(bins=15, fill=True))
plt.suptitle('Relationship between Sepal Length and Sepal Width', y=1.02)
plt.show()
```

```python
# Ques 14: dataset-- iris
#     Create a paiplot for the dataset to show relation between each pair of features based on different species.
#         ## output should be as shown below
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


iris = sns.load_dataset('iris')


sns.pairplot(iris, hue='species', palette='Set1', diag_kind='kde', markers=['o', 's', 'D'])
plt.suptitle('Pairplot of Iris Dataset by Species', y=1.02)
plt.show()
```

```python
# Ques 15: dataset-- exercise
#     Draw a pointplot to show pulse as a function of three categorical factors rest, walking and running.
#     ## output should be as shown below
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


exercise = sns.load_dataset('exercise')


plt.figure(figsize=(10, 6))
sns.pointplot(data=exercise, x='time', y='pulse', hue='kind', markers=['o', 's', 'D'], linestyles=['-', '--', '-.'], dodge=True)
plt.title('Pulse as a Function of Exercise Type')
plt.xlabel('Time (min)')
plt.ylabel('Pulse')
plt.legend(title='Exercise Type')
plt.grid(True)
plt.show()
```
